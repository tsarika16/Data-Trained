```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.heatmap(df.corr(), cmap = 'Blues', annot = True)

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
d = df[lambda df: df['Region'] == 'North America']
sns.heatmap(d.corr(), cmap = 'pink', annot = True)

plt.show()
```
